```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
output_notebook()
```

```python
opts = dict(plot_width=250, plot_height=250, min_border=0)
```

```python
# the comms handle repr show what cell it can be used to update
p1 = figure(**opts)
r1 = p1.circle([1,2,3], [4,5,6], size=20)

p2 = figure(**opts)
r2 = p2.circle([1,2,3], [4,5,6], size=20)

# get a handle to update the shown cell with
t = show(row(p1, p2), notebook_handle=True)
```

```python
import ipywidgets as widgets
colours = ['red', 'green', 'pink','blue','white']
r1.glyph.fill_color = colours[0]
push_notebook(t)
colourDD = widgets.Dropdown(options=colours, value=colours[0])

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        r1.glyph.fill_color = change['new']
        push_notebook(t)

colourDD.observe(on_change)

colourDD
```
